The Method I will use to reach my conclusion is simply Go over each statistic one by one and graphically trend each, including any interesting dates that may relate to each.

```python
# These are the tools I plan to use for this project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
plt.style.use('seaborn-whitegrid')
```

```python
# Let's get the data that I created to do our analysis.
statistics = pd.read_csv('USstatistics.csv')
statistics.info()
```

```python
# What the whole table looks like 
statistics.dropna()[:5]
```

### U.S. Population

```python
plt.figure(figsize = (10,6))
plt.plot(statistics['Year'],statistics['US Population'])
plt.title('US Population by Year')
plt.xlabel('Year')
plt.ylabel('US Population (in hundreds of millions)');
```

### Abortion

```python
statistics[['Year','Abortions','Abortion Rate']].dropna()[:5]
```

```python
plt.figure(figsize = (10,6))
plt.plot(statistics['Year'],statistics['Abortions'])
plt.xlim(1970,2020)
plt.ylim(0,2000000)
plt.vlines(1973,0,5000000,linestyles = 'dotted',color = 'r', label='Roe V. Wade (1973)')
plt.vlines(1991,0,5000000,linestyles = 'dotted',color = 'b', label ='Internet Becomes Public')
plt.title('Total Abortions Per Year')
plt.xlabel('Year from 1973 - 2019')
plt.ylabel('Abortions (in millions)')
plt.legend();
```

```python
plt.figure(figsize = (10,6))
plt.plot(statistics['Year'],statistics['Abortion Rate'])
plt.xlim(1970,2020)
plt.hlines(9.8,1970,2020,linestyles = 'solid',label='Abortion Rate of 1973')
plt.title('Abortion Rate')
plt.xlabel('Year')
plt.ylabel('# of Abortions/Population of Adult Women')
plt.legend();
```

Interestingly enough, the abortion rate per 1000 women is actually lower than it was when Roe V. Wade in 1973, I placed on the first graph when the internet became open to the public but it's difficult to say why the abortion rate has gone down so far.

## Teen Statistics

This includes statistics of potentially degrading teen morality such as:
1. Pregnancies/Birthrate
2. Smoking Rate
3. Marijuana Use
4. Cocaine Use

For 2,3, and 4, statistics start in the year 1980, while for 1, statistics start in 1950

```python
# First Few Statistics in table form
statistics[['Year','Teenage Birthrate','Teen Smoking','Teen Marijuana','Teen cocaine']].dropna()[:5]
```

Since these are all simply rates we can just put them all on the same graph.

```python
plt.figure(figsize=(10,6))
plt.xlim(1950,2020)
plt.ylim(0,100)
plt.plot(statistics['Year'],statistics['Teenage Birthrate'],color='blue',label='Teen Birthrate')
plt.plot(statistics['Year'],statistics['Teen Smoking'],color='red',label='Teen Smoking Rate')
plt.plot(statistics['Year'],statistics['Teen Marijuana'],color='green',label = 'Teen Marijuana Rate')
plt.plot(statistics['Year'],statistics['Teen cocaine'],color='black',label='Teen Cocaine Rate')
plt.vlines(1973,0,100,linestyles = 'dotted',color = 'r', label='Roe V. Wade (1973)')
plt.xlabel('Year')
plt.ylabel('Rate')
plt.title('Different Teen Related Metrics')
plt.legend();
```

## Crime Statistics

This section involves rates of crime over the years:
1. Violent Crime
2. Robbery
3. Aggravated Assault
4. Murder Rate

```python
statistics[['Year','Violent Crime','Robbery','Aggravated Assault','Murder Rate']].dropna()[:5]
```

```python
plt.figure(figsize=(10,6))
plt.xlim(1960,2020)
plt.plot(statistics['Year'],statistics['Violent Crime'],color='blue',label='Violent Crime')
plt.plot(statistics['Year'],statistics['Robbery'],color='red',label='Robbery')
plt.plot(statistics['Year'],statistics['Aggravated Assault'],color='green',label ='Aggravated Assault')
plt.plot(statistics['Year'],statistics['Murder Rate'],linestyle='dotted',color='black',label='Murder Rate')
plt.xlabel('Year')
plt.ylabel('Occurances per 100000')
plt.title('Crime in the US')
plt.legend();
```

```python
# Divorce Rate
statistics[['Year','Divorce Rate']].dropna()[:5]
```

```python
plt.figure(figsize=(10,6))
plt.xlim(1995,2020)
plt.ylim(0,5)
plt.plot(statistics['Year'],statistics['Divorce Rate'],marker='.');
plt.xlabel('Year')
plt.ylabel('Rate')
plt.title('Divorce Rate in the US');
```

```python
# Homelessness
# This could be more easily debated due to various data gathering methods
# So I will leave it out of the final conclusion
plt.figure(figsize=(10,6))
plt.plot(statistics['Year'],statistics['Homelessness'],marker='.');
plt.ylim(0,1000000)
plt.xlabel('Year')
plt.ylabel('Population of homeless (in millions)')
plt.title('Homelessness in US');
```

```python
# Poverty
plt.figure(figsize=(10,6))
plt.ylim(0,20)
plt.plot(statistics['Year'],statistics['Poverty Rate'],color='red',marker='.')
plt.xlabel('Year')
plt.ylabel('Poverty Rate')
plt.title('Poverty Rate in US');
```

```python
# Smoking Rate
plt.figure(figsize = (10,6))
plt.ylim(0,50)
plt.plot(statistics['Year'],statistics['Smoking Rate'],color='black',marker='.')
plt.xlabel('Year')
plt.ylabel('Percentage of Population')
plt.title('Percentage of U.S. Smoking Population');
```
